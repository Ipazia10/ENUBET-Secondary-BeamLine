# Decay Lenght for $K^{+}$ and $\pi^{+}$ at 6 - 7 - 8.5 GeV/c

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import scipy.optimize as opt
from scipy.stats import norm
```

### parameters: 

```python
tau_k = 1.2380*(pow(10,-8)) 
tau_pi = 2.6033*(pow(10,-8))
p = [6, 7, 8.5]
m_k = 0.493677 #GeV
m_pi = 0.13957018 #GeV 
c = 2.99*pow(10, 8)
```

### decay function: 

```python
def decay(x, p, tau, m):
     return np.exp(-(x*m)/(p*tau*c))
```

## $K^{+}$

```python
y = []
x = []
Pl = []
for mom in range (3): 
    y.append(np.array([]))
    x.append(np.array([]))
    Pl.append(np.array([]))
    for i in range(50):
        y[mom] = np.append(y[mom], decay(i, p[mom], tau_k, m_k))
        x[mom] = np.append(x[mom], i)
        
    print('P = '+str(p[mom]))
    
    Pl[mom] = plt.plot(x[mom],y[mom], '.-',markersize=12, label='P = '+str(p[mom]) + ' GeV/c')
    plt.xlabel("Distance [m]")
    plt.grid(True)
    plt.ylabel("Survival Rate [%]")
    #plt.axis([0, 40, 0.8, 1])
    plt.title("$K^{+}$ Decay length at 6 - 7 - 8.5 GeV/c")
    plt.legend()  
    plt.rcParams["figure.figsize"] = (10,20)
    #plt.figure(figsize=(20,10))
```
